On importe les données test et train et on affiche le contenu des DataFrames

```python
path = 'data_projet_rendre/train.csv'
path_test = 'data_projet_rendre/test.csv'

data_train = pd.read_csv(path)
data_test = pd.read_csv(path_test)

print(data_train.head())
print(data_test.head())
```

La boucle suivante sert à convertir la colonne cabine en 1's si l'individu a une cabine 0's sinon. (peut surement être amélioré si on regarde la lettre devant le numéro de cabine par exemple). La boucle sort un warning mais on peut l'ignorer, c'est vérifié.

```python
for i in range(0, len(data_train['Cabin'])):
    if type(data_train['Cabin'][i]) == float:
        data_train['Cabin'][i] = 0
    else:
        data_train['Cabin'][i] = 1
        
for i in range(0, len(data_test['Cabin'])):
    if type(data_test['Cabin'][i]) == float:
        data_test['Cabin'][i] = 0
    else:
        data_test['Cabin'][i] = 1
```

On vérifie la présence de NAs avec isna().sum(), on voit que certaines variables contiennent des NAs (Age, Embarked et Fare)

```python
print(data_train.isna().sum())
print(data_test.isna().sum())
```

Embarked n'est pas encore recodé en numérique, impossible de passer par une boucle et puis y'en a que deux donc on retire les NAs "à la main" (i.e. on localise l'indice et on supprime la ligne correspondante) NAs est une liste qui contient les indices de lignes contenant des NAs dans le DataFrame

```python
print(data_train['Embarked'])

NAs = [61, 829]
```

On retire les indices de la liste NAs du DataFrame et on fait une boucle pour Age. Idem; on créer une liste (missing) des indices de lignes contenant des NAs et on les supprime du DataFrame. A la différence qu'on le fait automatiquement avec une boucle ici

```python
data_train.drop(NAs, axis = 0, inplace = True)
data_train.reset_index(inplace = True)

missing = []

for i in range(0, len(data_train)):
    if np.isnan(data_train['Age'][i]):
        missing.append(i)
        
data_train.drop(missing, axis = 0, inplace = True)
data_train.reset_index(inplace = True, drop = True)
```

Manip similaire sur le jeu test MAIS: impossible de supprimer des lignes car le DataFrame doit absolument être (418, .). On est obligés de faire de l'imputation. 

Solution la plus simple: remplacer NAs par la moyenne arithmétique. Il existe de meilleures solutions à tester (MissForest ?) car remplacer par la moyenne c'est biaisé en terme d'échantillonage etc...

```python
mean_Age = np.mean(data_test['Age'])
mean_Fare = np.mean(data_test['Fare'])

for i in range(0, len(data_test)):
    if np.isnan(data_test['Age'][i]):
        data_test['Age'][i] = mean_Age
    elif np.isnan(data_test['Fare'][i]):
        data_test['Fare'][i] = mean_Fare
```

On affiche les BoxPlots des variables quanti pour détecter visuellement les outliers.

Plusieurs valeurs de Fare > 500 -> erreurs de saisie, donc on supprime si Fare >= 500. A voir si Age serait pas aussi à nettoyer ?

```python
outlier = (data_train[data_train['Fare'] >= 500].index.values)
outlier = outlier.tolist()

data_train.drop(outlier, axis = 0, inplace = True)
data_train.reset_index(drop = True, inplace = True)
```

On recode Sex en 1 si male et 0 si female

```python
for i in range(0, len(data_train['Sex'])):
    if data_train['Sex'][i] == 'male':
        data_train['Sex'][i] = 1
    else:
        data_train['Sex'][i] = 0
  
for i in range(0, len(data_test['Sex'])):
    if data_test['Sex'][i] == 'male':
        data_test['Sex'][i] = 1
    else:
        data_test['Sex'][i] = 0
```

On recode la variable Embarked en nombres entiers positifs (0 pour S, 1 pour Q et 2 pour C)

```python
for i in range(0, len(data_train['Embarked'])):
    if data_train['Embarked'][i] == 'S':
        data_train['Embarked'][i] = 0
    elif data_train['Embarked'][i] == 'Q':
        data_train['Embarked'][i] = 1
    else:
        data_train['Embarked'][i] = 2
        
for i in range(0, len(data_test['Embarked'])):
    if data_test['Embarked'][i] == 'S':
        data_test['Embarked'][i] = 0
    elif data_test['Embarked'][i] == 'Q':
        data_test['Embarked'][i] = 1
    else:
        data_test['Embarked'][i] = 2
```

On créer nos variables X et y. On passe les variable quali en entiers positifs (pas nécéssaire pour RF mais nécéssaire pour keras et ANN)

```python
X = data_train.drop(['Survived', 'Name', 'index', 'Ticket', 'PassengerId'], axis = 1)
y = data_train['Survived']

X_test = data_test.drop(['Name', 'Ticket', 'PassengerId'], axis = 1)

X['Sex'] = X['Sex'].astype(int)
X['Embarked'] = X['Embarked'].astype(int)
X['Cabin'] = X['Cabin'].astype(int)
X_test['Sex'] = X_test['Sex'].astype(int)
X_test['Embarked'] = X_test['Embarked'].astype(int)
X_test['Cabin'] = X_test['Cabin'].astype(int)
```

```python
# Logistic regression
logistic = LogisticRegression(max_iter = 1000, solver = 'newton-cg')
logistic.fit(X, y)

logistic_pred = logistic.predict(X_test)
```

Les 3 lignes ci dessous reviennent souvent: elles servent à concaténer PassengerId et les prédictions des modèles dans un DataFrame soumission, on peut ensuite écrire ce DataFrame en .csv avec le format nécéssaire pour Kaggle

```python
listes = list(zip(data_test['PassengerId'], logistic_pred))
soumission = pd.DataFrame(listes, columns = ['PassengerId', 'Survived'])
soumission.to_csv('submission.csv', header = True, index = False)
```

```python
# Single CART

tree = DecisionTreeClassifier(random_state = 1000,
                              min_samples_split = 2,
                              max_depth = None,
                              class_weight = 'balanced',
                              max_features = None)
tree.fit(X, y)

tree_pred  = tree.predict(X_test)
```

```python
listes = list(zip(data_test['PassengerId'], tree_pred))
soumission = pd.DataFrame(listes, columns = ['PassengerId', 'Survived'])
soumission.to_csv('submission.csv', header = True, index = False)
```

On a une moins bonne qualité de prédiction avec CART qu'avec regression logistique -> possiblement overfitting -> RF devrait régler le problème. 

A voir les scores déjà existants sur Kaggle, on sait qu'il est possible d'avoir une précision de 100%, du coup on peut laisser min_samples_split à 2

```python
# Random Forest
RF = RandomForestClassifier(n_estimators = 1000,
                            max_depth = None,
                            min_samples_split = 2,
                            class_weight = "balanced",
                            random_state = 1000,
                            max_features = None)

RF.fit(X, y)

RF_pred  = RF.predict(X_test)
```

```python
listes = list(zip(data_test['PassengerId'], RF_pred))
soumission = pd.DataFrame(listes, columns = ['PassengerId', 'Survived'])
soumission.to_csv('submission.csv', header = True, index = False)
```

```python
# Grid Search CV Random Forest
parameters = {'max_features' : [1, 2, 3, 4, 5, 6, 7, 8],
             'n_estimators' : [10, 100, 1000, 5000],
             'max_depth' : [None, 1, 5, 10, 20]}

grid_RF = RandomForestClassifier()
grid_RF = GridSearchCV(RF, parameters, scoring = 'f1_macro', cv = 10)

grid_RF.fit(X, y)

grid_RF_pred  = grid_RF.predict(X_test)
```

```python
listes = list(zip(data_test['PassengerId'], grid_RF_pred))
soumission = pd.DataFrame(listes, columns = ['PassengerId', 'Survived'])
soumission.to_csv('submission.csv', header = True, index = False)
```

```python
# Gradient Boosting

boost_RF = GradientBoostingClassifier(n_estimators = 1000,
                                     learning_rate = 0.1,
                                     max_depth = None,
                                     random_state = 1000)

boost_RF.fit(X, y)

boost_RF_pred  = boost_RF.predict(X_test)
```

```python
listes = list(zip(data_test['PassengerId'], boost_RF_pred))
soumission = pd.DataFrame(listes, columns = ['PassengerId', 'Survived'])
soumission.to_csv('submission.csv', header = True, index = False)
```

```python
# Neural Net
y = np.asarray(y)
X = np.asarray(X)
```

```python
# Neural Net

MLP = tf.keras.Sequential([
    tf.keras.layers.Dense(50, activation = 'relu'), # 1 hidden layer w/ 10 neurons
    tf.keras.layers.Dense(10, activation = 'relu'), # 1 hidden layer w/ 10 neurons
    tf.keras.layers.Dense(2)])

MLP.compile(optimizer = 'adam',
                  loss = tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
                  metrics = ['accuracy'])

MLP.fit(X, y, epochs = 10)

MLP_pred = MLP.predict(X_test)
```
